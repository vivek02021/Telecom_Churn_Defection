```python
#No : 73.5% and Yes=26.5% Data imbalance
```

```python
#separate all object type data and hold in df_cat
df_cat = df.select_dtypes("object")
```

```python
#sepeate all numeric datatype type data and hold in df_num
df_num = df.select_dtypes(["float64","int64"])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df_cat:
    #create object of LabelEncoder
    le=LabelEncoder()
    df_cat[col] = le.fit_transform(df_cat[col])
```

```python
df_cat.head()
```

```python
#to concatenate df_num,df_cat and hold in new data frame df_new
df_new = pd.concat([df_num,df_cat],axis=1)
```

```python
df_new.head()
```

```python
#separate input and output variable from df_new dataframe:
X=df_new.drop("Churn",axis=1)  #input variable
Y=df_new["Churn"]  #Target Variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#function create
def create_model(model):
    model.fit(X_train,Y_train)  #we train the model
    y_pred=model.predict(X_test) #test
    print(classification_report(Y_test,y_pred))
    return model
```

```python
from sklearn.metrics import classification_report
```

```python
'''There are 2 types of sampling technique :-
1. Over sampling tech. : If you are increasing the minority class that is
known as over sampling.
2. under sampling tech : If you have basically reduced the majority class
that is known as under sampling'''
```

```python
#!pip install imblearn
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
pd.Series(Y_train).value_counts()
```

```python
pd.Series(Y_test).value_counts()
```

```python
#create object of RandomOverSampler() class
ros = RandomOverSampler()
```

```python
X_sample2,y_sample2 = ros.fit_sample(X_train,Y_train) 
```

```python
pd.Series(y_sample2).value_counts()
```

```python
#we use DecisionTreeClassifier
dt2 = DecisionTreeClassifier(max_depth=3) #pruning tech.
```

```python
dt2.fit(X_sample2,y_sample2)  #we train the model
```

```python
X_sample_test,y_sample_test = ros.fit_sample(X_test,Y_test)
```

```python
y_pred = dt2.predict(X_sample_test)  #test the model
```

```python
print(classification_report(y_sample_test,y_pred))
```

```python
#create object of RandomUnderSampler() class
rus = RandomUnderSampler()
```

```python
X_sample1,y_sample1 = rus.fit_sample(X_train,Y_train) 
```

```python
pd.Series(y_sample1).value_counts()
```

```python
#we use DecisionTreeClassifier
d3 = DecisionTreeClassifier(max_depth=6)  #pruning tech.  not more than 8
```

```python
dt3.fit(X_sample1,y_sample1)  #we train the model
```

```python
y_pred = dt3.predict(X_test)  #test the model
```

```python
print(classification_report(Y_test,y_pred))
```
